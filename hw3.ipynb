# Q1: Solution

````python
def func1(names, db):
    idx = []
    for i in names:
        if i in db:
            idx.append(db.index(i))
    return idx
````

````python
names = list(range(10, 25000, 28))
db = list(range(1000000))
````

````python
%%time
print(func1(names, db))
````

````python
def func2(names, db):
    db = np.array(db)
    return db[np.isin(db, names)]
````

````python
%%time
print(func2(names, db)) #It is much faster by using numpy.isin method
````

````python
def search_pairs(coordinates, cirle_r0):
    diff = coordinates[:, np.newaxis, :] - coordinates #get the difference of the data by propagation
    diff_mat = np.linalg.norm(diff, axis = 2)
    keep = np.linalg.norm(coordinates, axis = 1) < cirle_r0
    new_mat = diff_mat[keep[:, np.newaxis] ^ keep] #filter those invalid points
    result = np.argsort(new_mat)[-6:] #Since the matrix is symmetric, we need to keep 6 results and then delete half of them later on
    filt = new_mat[result]
    ind_res = np.where(np.isin(diff_mat, filt)) #6 results are expected
    return np.vstack((ind_res[0], ind_res[1])).T[-3:] #The first 3 and the last 3 results are the same, so just keep one batch
````

````python
coordinates = np.array(
[[0.968, 0.313],
 [0.692, 0.876],
 [0.895, 0.085],
 [0.039, 0.17 ],
 [0.878, 0.098]]
)
cirle_r0 = 0.9
````

````python
print(search_pairs(coordinates, cirle_r0))
````
